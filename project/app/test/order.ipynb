```python
# ==============================
# –ò–ú–ü–û–†–¢–´ –ò –ù–ê–°–¢–†–û–ô–ö–ò
# ==============================
import requests
import json
from pprint import pprint

BASE_URL = "http://127.0.0.1:8000"
USERNAME = "admin"
PASSWORD = "admin"

print("üîê –ü–æ–ª—É—á–∞–µ–º —Ç–æ–∫–µ–Ω –∞–≤—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏...")

token_response = requests.post(
    f"{BASE_URL}/auth/token",
    data={"username": USERNAME, "password": PASSWORD},
)

if token_response.status_code != 200:
    raise Exception(f"–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å —Ç–æ–∫–µ–Ω: {token_response.text}")

token = token_response.json()["access_token"]
headers = {"Authorization": f"Bearer {token}"}

print("‚úÖ –¢–æ–∫–µ–Ω —É—Å–ø–µ—à–Ω–æ –ø–æ–ª—É—á–µ–Ω:", token[:25], "...\n")
```

```python
# ==============================
# –°–û–ó–î–ê–ù–ò–ï –ó–ê–ö–ê–ó–ê (POST /order)
# ==============================

new_order = {
    "date": "2025-10-20",
    "customer": "–ê–ª–µ–∫—Å–µ–π –ù–æ–≤–∏–∫–æ–≤",
    "phone": "+79998887766",
    "products": "–ö–æ–≥—Ç–µ—Ç–æ—á–∫–∞ (1), –î–æ–º–∏–∫ –¥–ª—è –∫–æ—à–∫–∏ (1)",
    "sum": "4590 —Ä—É–±",
    "status": "–ù–æ–≤—ã–π",
    "payment": "–ö–∞—Ä—Ç–∞ –æ–Ω–ª–∞–π–Ω",
    "delivery": "–ö—É—Ä—å–µ—Ä (–ú–æ—Å–∫–≤–∞)",
    "track": "TRK990001"
}

print("üì¶ –°–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π –∑–∞–∫–∞–∑...")

response = requests.post(
    f"{BASE_URL}/order/",
    headers=headers,
    json=new_order,
)

if response.status_code == 201:
    print("‚úÖ –ó–∞–∫–∞–∑ —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω!")
    pprint(response.json())
    order_id = response.json()["id"]
else:
    print(f"‚ùå –û—à–∏–±–∫–∞: {response.status_code}")
    print(response.text)
```

```python
# ==============================
# –ü–û–õ–£–ß–ï–ù–ò–ï –°–ü–ò–°–ö–ê –ó–ê–ö–ê–ó–û–í (GET /order)
# ==============================

print("üìã –ü–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–æ–∫ –∑–∞–∫–∞–∑–æ–≤...")

response = requests.get(f"{BASE_URL}/order/", headers=headers)

if response.status_code == 200:
    orders = response.json()
    print(f"‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω–æ –∑–∞–∫–∞–∑–æ–≤: {len(orders)}\n")
    pprint(orders)
else:
    print(f"‚ùå –û—à–∏–±–∫–∞: {response.status_code}")
    print(response.text)
```

```python
# ==============================
# –ü–û–õ–£–ß–ï–ù–ò–ï –û–î–ù–û–ì–û –ó–ê–ö–ê–ó–ê (GET /order/{id})
# ==============================

print(f"üîç –ü–æ–ª—É—á–∞–µ–º –∑–∞–∫–∞–∑ ID={order_id}...")

response = requests.get(f"{BASE_URL}/order/{order_id}", headers=headers)

if response.status_code == 200:
    print("‚úÖ –ó–∞–∫–∞–∑ –Ω–∞–π–¥–µ–Ω:")
    pprint(response.json())
else:
    print(f"‚ùå –û—à–∏–±–∫–∞: {response.status_code}")
    print(response.text)
```

```python
# ==============================
# –û–ë–ù–û–í–õ–ï–ù–ò–ï –ó–ê–ö–ê–ó–ê (PUT /order/{id})
# ==============================

update_data = {
    "status": "–°–æ–±—Ä–∞–Ω",
    "track": "TRK990001-UPDATED"
}

print(f"‚úèÔ∏è –û–±–Ω–æ–≤–ª—è–µ–º –∑–∞–∫–∞–∑ ID={order_id}...")

response = requests.put(
    f"{BASE_URL}/order/{order_id}",
    headers=headers,
    json=update_data,
)

if response.status_code == 200:
    print("‚úÖ –ó–∞–∫–∞–∑ —É—Å–ø–µ—à–Ω–æ –æ–±–Ω–æ–≤–ª—ë–Ω:")
    pprint(response.json())
else:
    print(f"‚ùå –û—à–∏–±–∫–∞: {response.status_code}")
    print(response.text)
```

```python
# ==============================
# –£–î–ê–õ–ï–ù–ò–ï –ó–ê–ö–ê–ó–ê (DELETE /order/{id})
# ==============================

print(f"üóëÔ∏è –£–¥–∞–ª—è–µ–º –∑–∞–∫–∞–∑ ID={order_id}...")

response = requests.delete(f"{BASE_URL}/order/{order_id}", headers=headers)

if response.status_code == 204:
    print("‚úÖ –ó–∞–∫–∞–∑ —É—Å–ø–µ—à–Ω–æ —É–¥–∞–ª—ë–Ω.")
else:
    print(f"‚ùå –û—à–∏–±–∫–∞: {response.status_code}")
    print(response.text)
```
