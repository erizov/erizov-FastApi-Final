```python
# –ò–º–ø–æ—Ä—Ç –∏ –±–∞–∑–æ–≤—ã–π URL

import requests
import json
from pprint import pprint

BASE_URL = "http://127.0.0.1:8000"

# –ü–æ–ª—É—á–∞–µ–º —Ç–æ–∫–µ–Ω
USERNAME = "admin"
PASSWORD = "admin"
token_response = requests.post(
    f"{BASE_URL}/auth/token",
    data={"username": USERNAME, "password": PASSWORD}
)

if token_response.status_code != 200:
    raise Exception(f"–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å —Ç–æ–∫–µ–Ω: {token_response.text}")

token = token_response.json()["access_token"]
headers = {"Authorization": f"Bearer {token}"}

print("‚úÖ –¢–æ–∫–µ–Ω –ø–æ–ª—É—á–µ–Ω:", token[:20], "...")  # –≤—ã–≤–æ–¥–∏–º –ø–µ—Ä–≤—ã–µ 20 —Å–∏–º–≤–æ–ª–æ–≤
```

```python
# –í—ã–±–æ—Ä ID –∫–ª–∏–µ–Ω—Ç–∞
client_id = 1

# –û—á–∏—Å—Ç–∫–∞ –∏—Å—Ç–æ—Ä–∏–∏ –¥–∏–∞–ª–æ–≥–∞
resp = requests.post(
    f"{BASE_URL}/dialog/clear",
    json={"client_id": client_id},
    headers=headers,
)
if resp.status_code == 200:
    print(resp.json()["message"])
else:
    print(f"–û—à–∏–±–∫–∞ –æ—á–∏—Å—Ç–∫–∏ –ø–µ—Ä–µ–¥ –∑–∞–ø—É—Å–∫–æ–º: {resp.status_code} {resp.text}")
    
# –§—É–Ω–∫—Ü–∏—è –æ—Ç–ø—Ä–∞–≤–∫–∏ —Å–æ–æ–±—â–µ–Ω–∏—è
def send_message(user_input):
    print(f'–í—ã: {user_input}')
    resp = requests.post(f"{BASE_URL}/dialog/request", json={"user_input": user_input, "client_id": client_id}, headers=headers)
    data = resp.json()
    print("–ê—Å—Å–∏—Å—Ç–µ–Ω—Ç:")
    pprint(data["response"])    
    
# –î–∏–∞–ª–æ–≥
def dialog():
    """
    –ó–∞–ø—É—Å–∫–∞–µ—Ç –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–π —Ü–∏–∫–ª –¥–∏–∞–ª–æ–≥–∞ —Å –∞—Å—Å–∏—Å—Ç–µ–Ω—Ç–æ–º.
    –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –≤–≤–æ–¥–∏—Ç —Å–æ–æ–±—â–µ–Ω–∏—è –≤ –∫–æ–Ω—Å–æ–ª—å,
    –ø–æ–∫–∞ –Ω–µ –≤–≤–µ–¥—ë—Ç 'stop' –∏–ª–∏ 'quit'.
    """
    print("üí¨ –í–≤–µ–¥–∏—Ç–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –∞—Å—Å–∏—Å—Ç–µ–Ω—Ç–∞ (–∏–ª–∏ 'stop' / 'quit' –¥–ª—è –≤—ã—Ö–æ–¥–∞):\n")

    while True:
        user_input = input("–í—ã: ").strip()
        if user_input.lower() in {"stop", "quit"}:
            print("\nüõë –î–∏–∞–ª–æ–≥ –∑–∞–≤–µ—Ä—à—ë–Ω –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º.")
            break

        try:
            send_message(user_input)
        except Exception as e:
            print(f"‚ö†Ô∏è –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ—Ç–ø—Ä–∞–≤–∫–µ —Å–æ–æ–±—â–µ–Ω–∏—è: {e}")    
```

```python
# –ó–∞–ø—É—Å–∫ –¥–∏–∞–ª–æ–≥–∞
dialog()
```
