```python
import requests
import json
from pprint import pprint

# ==============================
# –ë–ê–ó–û–í–´–ï –ù–ê–°–¢–†–û–ô–ö–ò
# ==============================
BASE_URL = "http://127.0.0.1:8000"
USERNAME = "admin"
PASSWORD = "admin"

# –ü–æ–ª—É—á–∞–µ–º JWT —Ç–æ–∫–µ–Ω
print("üîê –ü–æ–ª—É—á–∞–µ–º —Ç–æ–∫–µ–Ω –∞–≤—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏...")
token_response = requests.post(
    f"{BASE_URL}/auth/token",
    data={"username": USERNAME, "password": PASSWORD},
)

if token_response.status_code != 200:
    raise Exception(f"–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–ª—É—á–∏—Ç—å —Ç–æ–∫–µ–Ω: {token_response.text}")

token = token_response.json()["access_token"]
headers = {"Authorization": f"Bearer {token}"}

print("‚úÖ –¢–æ–∫–µ–Ω —É—Å–ø–µ—à–Ω–æ –ø–æ–ª—É—á–µ–Ω:", token[:25], "...\n")
```

```python
# –¢–µ—Å—Ç —á—Ç–µ–Ω–∏—è –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π (GET /faq)
print("üìò –¢–µ—Å—Ç: —á—Ç–µ–Ω–∏–µ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π (GET /faq)")
response = requests.get(f"{BASE_URL}/base/faq", headers=headers)
print("–°—Ç–∞—Ç—É—Å:", response.status_code)
if response.status_code == 200:
    data = response.json()
    pprint(data['text'][:1000])
elif response.status_code == 404:
    print("‚ö†Ô∏è  –§–∞–π–ª –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π –Ω–µ –Ω–∞–π–¥–µ–Ω")
else:
    print("‚ùå –û—à–∏–±–∫–∞:", response.text)
```

```python
# –¢–µ—Å—Ç –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è —Ñ–∞–π–ª–∞ (PUT /faq)
print("‚ôªÔ∏è –¢–µ—Å—Ç: –ø–µ—Ä–µ–∑–∞–ø–∏—Å—å —Ñ–∞–π–ª–∞ –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π (PUT /faq)")

updated_text = """
–§–∞–π–ª —É—Å–ø–µ—à–Ω–æ –æ–±–Ω–æ–≤–ª—ë–Ω!
–≠—Ç–æ –Ω–æ–≤–∞—è –≤–µ—Ä—Å–∏—è –±–∞–∑—ã –∑–Ω–∞–Ω–∏–π, —Å–æ—Ö—Ä–∞–Ω—ë–Ω–Ω–∞—è —á–µ—Ä–µ–∑ PUT.
"""

payload = {"text": updated_text}
response = requests.put(f"{BASE_URL}/base/faq", headers=headers, json=payload)
print("–°—Ç–∞—Ç—É—Å:", response.status_code)
try:
    pprint(response.json())
except Exception:
    print(response.text)
```
